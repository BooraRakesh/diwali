```python
import pandas as pd

# Try reading the file with a different encoding, like 'latin-1'
diwali_df = pd.read_csv('/content/drive/MyDrive/Diwali Sales Data.csv', encoding='latin-1')
```

```python
display (diwali_df)
```

```python
import plotly.express as px

fig = px.histogram(diwali_df, x='Gender')
fig.show()
```

```python
#2- for male and female give me sum of amount wrt each other and plot graph
gender_amount_sum = diwali_df.groupby('Gender')['Amount'].sum()
import plotly.express as px

fig = px.bar(gender_amount_sum, x=gender_amount_sum.index, y=gender_amount_sum.values,
             labels={'x':'Gender', 'y':'Sum of Amount'},
             title='Sum of Amounts by Gender')
fig.show()
```

```python
#3-count of each age group with individual count of age with respect to gender
age_gender_counts = diwali_df.pivot_table(index='Age Group', columns='Gender', values='Age', aggfunc='count')
print(age_gender_counts)
```

```python
#4- plot the total amount spent by each age group
#4- plot the total amount spent by each age group
import plotly.express as px

# Calculate the total amount spent by each age group
age_group_amount_sum = diwali_df.groupby('Age Group')['Amount'].sum()

fig = px.bar(age_group_amount_sum, x=age_group_amount_sum.index, y=age_group_amount_sum.values,
             labels={'x':'Age Group', 'y':'Total Amount Spent'},
             title='Total Amount Spent by Each Age Group')
fig.show()
```

```python
#5 plot the graph that depicts total number of orders from top 10 states

import pandas as pd
import plotly.express as px

# Assuming 'diwali_df' is already loaded in your environment
# If not, load your data like this:
# diwali_df = pd.read_csv('correct_path_to_your_data.csv')

# Calculate the total number of orders for each state
state_orders = diwali_df['State'].value_counts().head(10)

# Create a bar plot using Plotly Express
fig = px.bar(state_orders,
             x=state_orders.index,
             y=state_orders.values,
             labels={'x':'State', 'y':'Number of Orders'},
             title='Top 10 States by Number of Orders')
fig.show()
```

```python
#6 total amount of top 10 states
import pandas as pd

# Assuming 'diwali_df' is already loaded in your environment
# If not, load your data like this:
# diwali_df = pd.read_csv('correct_path_to_your_data.csv')

# Calculate the total amount for each state
state_amount_sum = diwali_df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)

print(state_amount_sum)
```

```python
#7 plot me a graph that show comparison between no of married and unmarried people
import pandas as pd
import plotly.express as px

# Assuming 'diwali_df' is already loaded in your environment
# If not, load your data like this:
# diwali_df = pd.read_csv('correct_path_to_your_data.csv')

# Count married and unmarried people
marital_status_counts = diwali_df['Marital_Status'].value_counts()

# Create a bar plot
fig = px.bar(marital_status_counts,
             x=marital_status_counts.index,
             y=marital_status_counts.values,
             labels={'x': 'Marital Status', 'y': 'Number of People'},
             title='Comparison of Married and Unmarried People',
             colour red)
fig.show()
```

```python
#8 plot me amount spent by each male and each female with respsct to marital status
import plotly.express as px

# Group the data by Gender and Marital_Status and calculate the total amount spent
gender_marital_amount = diwali_df.groupby(['Gender', 'Marital_Status'])['Amount'].sum().reset_index()

# Create the bar plot
fig = px.bar(gender_marital_amount, x='Gender', y='Amount', color='Marital_Status',
             barmode='group',
             labels={'Amount':'Total Amount Spent'},
             title='Amount Spent by Gender and Marital Status')
fig.show()
```

```python
#11 plot a graph that provides statistics on the intrest of customers with respect to product category
product_category_counts = diwali_df['Product_Category'].value_counts()
import plotly.express as px
fig = px.bar(product_category_counts,
             x=product_category_counts.index,
             y=product_category_counts.values,
             labels={'x': 'Product Category', 'y': 'Number of Customers Interested'},
             title='Customer Interest by Product Category')
fig.show()
```

```python
#12 i would like to know the budget spent on each product category in decending order
category_budget = diwali_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
print(category_budget)
```
